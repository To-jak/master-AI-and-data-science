```python
# Importing useful modules and packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import autocorrelation_plot
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from scipy.stats import norm
import warnings
warnings.filterwarnings('ignore')
```

# Internet traffic data

## Importing data

```python
df_lbl_tcp = pd.read_csv('lbl-tcp-3.tcp.txt', sep=' ', header=None, names=['timeStamp',
                                                                         'source host',
                                                                         'destination host',
                                                                         'source TCP port',
                                                                         'destination TCP port',
                                                                         'number of data bytes'])
df_lbl_tcp.head()
```

## Creating Time Series  
  
Each observation should corresponds to the number of packets transmitted within 10 seconds. We first create the time values for splitting the data in 10sec interval batch. The observation is made over a two hours period. That corresponds to 7200 secondes.

```python
ten_sec_time_values = np.arange(0, 7200, 10)
```

Then we keep for the observations all the data that falls into each 10 interval, and sum all the data bytes for each observation.

```python
observations = []
for t in ten_sec_time_values:
    ten_sec_group = df_lbl_tcp[(df_lbl_tcp['timeStamp'] > t ) & (df_lbl_tcp['timeStamp'] < t+10)]
    observations.append(sum(ten_sec_group['number of data bytes']))
    
print('Number of observations: {}'.format(len(observations)))
```

We obtain 720 observations as desired.

## Time series plot

```python
# Time series plot
plt.figure(figsize=(15,8))
plt.title('Time serie: number of data bytes for each 10 sec interval')
plt.plot(observations, label='Number of data bytes')
plt.xlabel('10 sec interval')
plt.show()
```

## Log transformation
  
A log transformation is applied for stabilizing the variance.

```python
# Log transformation
observations = np.log(observations)

# Time series plot
plt.figure(figsize=(15,8))
plt.title('Time serie: log(number of data bytes) for each 10 sec interval')
plt.plot(observations, label='Number of data bytes')
plt.xlabel('10 sec interval')
plt.show()
```

## Autoregressive model
We will try to fit an auto regressive model to the data. 

```python
plt.figure(figsize=(15,8))
autocorrelation_plot(observations)
plt.show()
```

The auto correlation plot suggests that we cloud try values from 1 to 10 for the p parameter of the AR(p) model.

## P value selection for the AR model
  
3 criterions will be observed for selecting the model:
* AIC
* BIC
* MSE

```python
# p values to test for selecting our AR model
p_values = np.arange(1,10)

AIC = []
BIC = []
MSEmean = []

# Getting AIC, BIC and computing MSE for each p value
for p in p_values:
    # Fitting ARIMA model with p value
    model = ARIMA(observations, order=[p,0,0])
    model_fit = model.fit(disp=0)
    # AIC
    AIC.append(model_fit.aic)
    # BIC
    BIC.append(model_fit.bic)   
    # MSE obtained with cross validation
    MSE = []
    n_split = 5
    tscv = TimeSeriesSplit(n_splits=5)
    for train_index, test_index in tscv.split(observations):
        model_fit = ARIMA(observations[train_index], order=[p,0,0]).fit()
        y_pred = model_fit.predict(start=test_index[0], end=test_index[-1])
        MSE.append(mean_squared_error(observations[test_index], y_pred))
    # mean MSE after cross validation
    MSEmean.append(np.mean(MSE))

# AIC & BIC plot
plt.figure()
plt.title('AIC and BIC according to p.')
plt.plot(p_values, AIC, 'b', label='AIC')
plt.plot(p_values, BIC, 'r', label='BIC')
plt.xlabel('p values')
plt.legend()
plt.show()

# MSE plot
plt.figure()
plt.title('Mean squarred prediction error according to p.')
plt.plot(p_values, MSEmean, label='MSE')
plt.xlabel('p values')
plt.legend()
plt.show()
```

**Observations**: AIC and BIC are the lowest for p values respectively equal to 4 and 3.  
Mean squarred prediction error is the lowest for a p value equal to 3.  
 
**For our AR(p) model, we will then take p=3 and then perform a AR(3) model.**

## AR(3) model

With the results obtained in the previous question, we choose to perform a AR(3) model on our observations.

```python
# Fitting the AR(3) model
model = ARIMA(observations, order=(3,0,0))
model_fit = model.fit()

# Model results summary
print(model_fit.summary())

# Getting residuals
residuals = pd.DataFrame(model_fit.resid, columns=['Residuals'])

# plot residuals
residuals.plot(figsize=(10,6))
plt.show()

# Residuals description
print(residuals.describe())

# residual estimated distribution plot
ax = residuals.plot(kind='kde')

# normal distribution with estimated mean and std
x = np.linspace(-3,4,500)
y = norm.pdf(x, loc=residuals.mean(), scale=residuals.std())
ax.plot(x,y, label='Normal distribution')
plt.legend()
plt.show()
```

**Observations**: We get a plot of the residual errors, and we see that the variance doesn't seem to be perfectly constant over time. It may correspond to some information that is not captured by the AR(3) model. It could be better. On the other hand, residuals are pretty much centered (mean almost equals 0, there is no biais in the model prediction) and their distribution almost fit a gaussian distribution.
